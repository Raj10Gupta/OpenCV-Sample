```python
import cv2

image = cv2.imread(r"C:\Users\rj100\OneDrive\Pictures\box3.jpg")
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,200)

_,contours,hierarchy = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

number_of_objects_in_image = len(contours)

print('The number is', str(number_of_objects_in_image))
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```
